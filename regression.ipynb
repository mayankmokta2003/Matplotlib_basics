```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

### lmplot()
regression plot is used in machine learning so we will not fully discuss what, how and where it is used but we will see how it is created and what are the values we can put inside it

```python
tips = sns.load_dataset("tips")
tips
```

```python
# suppose we want to predict tip through total_bill we will mark tip in y-axis and total_bill in x-axis
sns.lmplot(data=tips,x="total_bill",y="tip")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",palette="coolwarm")
```

```python
# through markers we made one dot as o other as v shaped and through scatter_kws we can change the size of shapes
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",palette="coolwarm",markers=["o","v"],scatter_kws={"s":100})
```
