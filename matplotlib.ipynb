```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

### MATPLOTLIB BASICS

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("Basic Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
```

### SUBPLOTS IN MATPLOTLIB

```python
x = np.array([1,2,3,4,5,6])
y = x

plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.plot(x,y**2)

plt.subplot(2,2,3)
plt.plot(x,y**3)

plt.subplot(2,2,4)
plt.plot(x,y//3)
```

### OBJECT ORIENTED WAY

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
fig2 = plt.figure()
axis2 = fig.add_axes([0.3,0.3,0.2,0.2])
axis2.plot(x,y)
```

### TYPES OF PLOTS

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```

### SAVING PLOTS

```python
# dpi just increases the visual quality of the graph 
fig = plt.figure(figsize=(9,5),dpi=200)
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis1.grid(True)
plt.savefig("basicplot.png")
```

### WORKING WITH IMAGES

```python
import matplotlib.image as mpimg

img = mpimg.imread("ex.png")
plt.imshow(img)
# to remove axis from image
plt.axis(False)
```
