```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

##### We use distribution plots to visualise the distribution of quantitative data. The data which can be different for almost everyone like age or salary

##### Some plots that allow us to visualise the distribution of data set. The plots are:
##### 1. distplot/histplot   imp
##### 2. jointplot   imp
##### 3. pairplot    imp
##### 4. rugplot 


```python
# thats how we import data from somewhere this tips in in github of seaborn but its no working
# df = sns.load_dataset("tips")

data = {
"total_bill":[16.99,10.34,21.01,23.68,24.59,25.29,8.77,26.88,15.04,14.78],
"tip":[1.01,1.66,3.50,3.31,3.61,4.71,2.00,3.12,1.96,3.23],
"sex":["Female","Male","Male","Male","Female","Male","Male","Male","Male","Female"],
"smoker":["No","Yes","No","No","Yes","No","No","Yes","No","No"],
"day":["Mon","Mon","Sun","Sun","Sun","Sun","Sat","Sun","Wed","Sun"],
"time":["Dinner","Dinner","Dinner","Dinner","Dinner","Dinner","Dinner","Dinner","Dinner","Dinner"],
"size":[2,3,3,2,1,4,2,5,6,2]
}

df = pd.DataFrame(data)

df
```

```python
# this makes histogram we can use sns.distplot as well and this bins will make cols as of number 
# if we make kde i.e kernel densitiy estimater True it will make a line through these cols
sns.histplot(df["total_bill"],bins=6, kde = True)
```

```python
# we can make subplots in histogram as well
plt.subplot(2,2,1)
sns.histplot(df["total_bill"],bins=6, kde = True)

plt.subplot(2,2,2)
sns.histplot(df["tip"],bins=6, kde = True)
```

#### JOINTPLOT

```python
# suppose there are values in x = 10 and y = 5 axis so it will create a dot bw points simple
# kind is used to make graphs suppose we write scatter it is default one
sns.jointplot(x="total_bill", y="tip", data=df, kind="scatter")
```

```python
sns.jointplot(x="total_bill", y="tip", data=df, kind="reg")
```

### PAIRPLOT

```python
# to make simple pairplot we just pass name eg df thats it
# suppose we want to categorize these all based on something eg gender we use hue
sns.pairplot(df,hue="sex")
```

```python

# we use palette to change colours for differences values
sns.pairplot(df,hue="size",palette="rainbow")
```

### RUGPLOT

```python
# here we can put whole dataframe or just one value

sns.rugplot(df["total_bill"])
```
